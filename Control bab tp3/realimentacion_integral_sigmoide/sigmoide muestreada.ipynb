```python
import numpy as np
import matplotlib.pyplot as plt
n=20
x=np.linspace(-1,1,n)
x=np.array([-1 ,-0.8, -0.65, -0.54,-0.45,-0.38,-0.30,-0.25,-0.2,-0.15,-0.1,-0.05,0,0.08,0.16,0.25,0.35,0.50,0.75,1])
y=1/(1+np.exp(-5*x-1))
```

```python
plt.figure()
plt.scatter(y=y,x=x)
plt.show()
```
